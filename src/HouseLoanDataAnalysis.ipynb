```python
X= loan_data.loc[:, loan_data.columns != 'TARGET']
y = loan_data.TARGET
```

### Encoding the columns

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import warnings
warnings.filterwarnings('ignore')
```

```python
X.loc[:,:] = X.loc[:,:].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.75,random_state=10)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
#X_train_smote,y_train_smote = smote.fit_resample(X_train.astype('float'),y_train)
X_smote,y_smote = smote.fit_resample(X.astype('float'),y)
```

```python
X_train_1,X_test_1,y_train_1,y_test_1 = train_test_split(X_smote,y_smote,train_size=0.75,random_state=10)
```

### Plotting Imbalance and Balance

```python
#y_train_smote.value_counts().plot(kind='bar')
y_train_1.value_counts().plot(kind='bar')
plt.title('Balanced TARGET column')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn1=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
#knn1.fit(X_train_smote,y_train_smote)
knn1.fit(X_train_1,y_train_1)
```

```python
pred = knn1.predict(X_test_1)
#y_pred_proba = rf.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score, roc_curve
cm = confusion_matrix(y_test_1,pred )
cm
```

### Prediction Model and ROC curve

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

print(X_train_1.shape[0])
print(y_train_1.shape)

model =  tf.keras.models.Sequential([
    tf.keras.layers.Input(shape= (121,)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=  'relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(64, activation=  'relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(10, activation = 'sigmoid')
])
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.summary()
#model.fit(X_train_smote, y_train_smote, epochs = 5)
    
```

```python
model.fit(X_train_1, y_train_1, epochs = 5)
```

```python
pred = model.predict(X_test_1)
results =  np.argmax(pred, axis =1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_1, results))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
clf1 = DecisionTreeClassifier()
clf2 = LogisticRegression();
```

```python
clf1.fit(X_train_1, y_train_1)
clf2.fit(X_train_1, y_train_1)

y_score1 = clf1.predict_proba(X_test_1)[:,1]
y_score2 = clf2.predict_proba(X_test_1)[:,1]
```

```python
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test_1, y_score1)
false_positive_rate2, true_positive_rate2, threshold2 = roc_curve(y_test_1, y_score2)
print('roc_auc_score for DecisionTree: ', roc_auc_score(y_test_1, y_score1)) 
print('roc_auc_score for Logistic Regression: ', roc_auc_score(y_test_1, y_score2))
```

```python
# Plot ROC curves
plt.subplots(1, figsize=(10,10))    
plt.title('Receiver Operating Characteristic - DecisionTree')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

plt.subplots(1, figsize=(10,10))    
plt.title('Receiver Operating Characteristic - Logistic regression')
plt.plot(false_positive_rate2, true_positive_rate2)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
